```python
class Person:
    def __init__(self, hair, eyes, skin, height: int, weight: int):
        self.hair = hair
        self.eyes = eyes 
        self.height = height
        self.weight = weight
        self.skin = skin
    
    def eat(self):
        return f"Person that weights at {self.weight} with {self.hair} hair now eats"
        
    def sleep(self):
        print(f"person with {self.hair} hair now sleeps because their {self.eyes} eyes are now closed")
        
    def walk(self):
        print(f"person with {self.skin} skin now walks and the height when walking is {self.height}")
        
    def play(self):
        print(f"the person with {self.hair} hair now plays")
    
    def study(self):
        print(f"the person is with {self.hair} hair is reading astrophysics with their {self.eyes} eyes")
    
#making a new actual person
Alex = Person("black", "green", "white", 180, 80)   

behaviours = ["eat", "sleep", "walk", "play", "study"]

for i in behaviours:
    method = getattr(Alex, i)
    method()
    print( "moments later")

Person.eat(Alex)


```

```python
class Administration:
    def __init__ (self, school, program, name, accept):
        self.school = school
        self. program = program
        self.name = name
        self.accept = accept

    def __str__(self):
        return f"{self.name} has applied to {self.school} and did he get accepted ?: accept:{self.accept}"
    
    def __repr__(self):
        return f"{self.name} has tried to get in {self.school} and actually choose {self.program}, but did he get accepted? \n accept: {self.accept}"
    

student1 = Administration("kth", "economy", "ANNA", True)

print(student1)
student1
```
